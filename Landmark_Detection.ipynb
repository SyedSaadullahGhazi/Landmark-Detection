```python
# Training of Model
from sklearn.preprocessing import LabelEncoder
lencoder=LabelEncoder()
lencoder.fit(df["landmark_id"])
```

```python
df.head()
```

```python
def encode_label(lbl):
    return lencoder.transform(lbl)  
```

```python
def decode_label(lbl):
    return lencoder.inverse_transform(lbl)
```

```python
def get_image_from_number(num, df):
        fname, label = df.iloc[num, :]
        fname = fname + '.jpg'
        f1 = fname[0]
        f2 = fname[1]
        f3 = fname[2]
        path = os.path.join(base_path,f1,f2,f3,fname)
        if not os.path.exists(path):  # Check if the image exists
            print(f"Image not found at {path}")
            return None, label
        im = cv2.imread(path)
        #im = cv2.imread(os.path.join(base_path, path))
        if im is None:
            print(f"Failed to read image from {path}")
        return im, label
```

```python
print("4 Sample images from random classes")
fig=plt.figure(figsize=(16,16))
for i in range(1,5):
    ri=random.choices(os.listdir(base_path), k=3)
    folder=base_path + '/' + ri[0] + '/' + ri[1] + '/' + ri[2]
    random_img=random.choice(os.listdir(folder))
    img=np.array(Image.open(folder+'/'+random_img))
    fig.add_subplot(1,4,i)
    plt.imshow(img)
    plt.axis('off')
plt.show()
```

```python
from keras.applications.vgg19 import VGG19
from keras.layers import *
from keras import Sequential
```

```python
# Parameters
learning_rate=0.0001
decay_speed=1e-6
momentum=0.09
loss_function="sparse_categorical_crossentropy"
source_model=VGG19(weights=None)
drop_layer=Dropout(0.5)
drop_layer2=Dropout(0.5)
```

```python
model=Sequential()
for layer in source_model.layers[:-1]:
    if layer==source_model.layers[-25]:
        model.add(BatchNormalization())
    model.add(layer)
model.add(Dense(num_classes,activation='softmax'))
model.summary()
```

```python
from tensorflow.keras.optimizers import RMSprop
optim1 = RMSprop(learning_rate=learning_rate)
model.compile(optimizer=optim1, loss=loss_function, metrics=["accuracy"])
```

```python
def image_reshape(im, target_size):
    if im is None:
        print("Error: Image is empty, cannot resize.")
        return None  
    # Resize the image if it's valid
    return cv2.resize(im, target_size)
```

```python
def get_batch(dataframe, start, batch_size):
    image_array = []  # Initialize as an empty list, not with "id"
    label_array = []  # Initialize as an empty list, not with "landmark_id"
    end_img = start + batch_size
    if end_img > len(dataframe):
        end_img = len(dataframe)
    for idx in range(start, end_img):
        n = idx
        im, label = get_image_from_number(n,dataframe)
        if im is None:
            print(f"Warning: Image {n} could not be loaded. Skipping this sample.")
            continue  # Skip this iteration if the image is None
        # Resize and normalize the image
        im = image_reshape(im, (224, 224)) / 255.0
        image_array.append(im)
        label_array.append(label)
    label_array = encode_label(label_array)
    return np.array(image_array), np.array(label_array)
```

```python
# Define parameters
batch_size = 16
epoch_shuffle = True
weight_classes = True
epochs = 1
train, val = np.split(df.sample(frac=1, random_state=42), [int(0.8 * len(df))])
print("Training set size:", len(train))
print("Validation set size:", len(val))
```

```python
for e in range(epochs):
    print("Epoch: "+str(e+1)+'/'+str(epochs))
    if epoch_shuffle:
        train=train.sample(frac=1)
    for it in range(int(np.ceil(len(train)/batch_size))):
        x_train,y_train=get_batch(train,it*batch_size,batch_size)
        model.fit(x_train,y_train)
model.save("Model.keras")
```
